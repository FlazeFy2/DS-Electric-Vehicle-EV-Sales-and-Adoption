```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Bar Chart Top 10 Most Sold Units per Brand Model
df = pd.read_csv("train.csv")

# Dataset
df["Brand Model"] = df["Brand"] + " " + df["Model"]

# Group & Sum Unit Sold
units_sold_per_model = df.groupby("Brand Model")["Units_Sold"].sum().sort_values(ascending=False).head(10)

# Plot Bar Chart
plt.figure(figsize=(12, 6))
ax = units_sold_per_model.plot(kind='bar')
# Label
for i, value in enumerate(units_sold_per_model):
    plt.text(i, value + 1, str(int(value)), ha='center', va='bottom')
plt.title("Top 10 Most Sold Units per Brand Model")
plt.xlabel("Brand Model")
plt.ylabel("Total Units Sold")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Bar Chart Top 10 Most Revenue per Brand Model
df = pd.read_csv("train.csv")

# Dataset
df["Brand Model"] = df["Brand"] + " " + df["Model"]

# Group & Sum Unit Sold
revenue_per_model = df.groupby("Brand Model")["Revenue"].sum().sort_values(ascending=False).head(10)

# Plot Bar Chart
plt.figure(figsize=(12, 6))
ax = revenue_per_model.plot(kind='bar')
# Label
for i, value in enumerate(revenue_per_model):
    plt.text(i, value + 1, f"{value:,.0f}", ha='center', va='bottom')
plt.title("Top 10 Most Revenue per Brand Model")
plt.xlabel("Brand Model")
plt.ylabel("Revenue Amount")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
