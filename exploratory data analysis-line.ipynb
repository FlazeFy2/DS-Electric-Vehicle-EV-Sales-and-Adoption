```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Unit Sold Per Month in 2023
data = pd.read_csv("train.csv")

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['YearMonth'] = data['Date'].dt.to_period('M').astype(str)

# Group by YearMonth and sum Unit Sold
monthly_sales = data.groupby('YearMonth')['Units_Sold'].sum()

# Plot Line Chart
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Unit Sold Per Month in 2023')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.grid(True)
# Add Labels
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y + 1, str(int(y)), ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Units Sold Per Quarter (Top 5 Brands)
data = pd.read_csv("train.csv")

# Convert Date column
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Extract Quarter-Year
data['Quarter'] = 'Q' + data['Date'].dt.quarter.astype(str) + '-' + data['Date'].dt.year.astype(str)

# Convert to numeric
data['Units_Sold'] = pd.to_numeric(data['Units_Sold'], errors='coerce')

# Remove NaN
data = data.dropna(subset=['Units_Sold'])

# Find Top 5 Brands by Total Units Sold
top5_brand = data.groupby('Brand')['Units_Sold'].sum().nlargest(5).index

# Filter dataset to Top 5 Brands
data = data[data['Brand'].isin(top5_brand)]

# Group by Brand and Quarter then calculate Total Units Sold
quarterly_sales = data.groupby(['Brand','Quarter'])['Units_Sold'].sum().reset_index()

# Pivot table for multi-line chart
pivot_sales = quarterly_sales.pivot(index='Quarter', columns='Brand', values='Units_Sold')

# Sort Quarter order
pivot_sales = pivot_sales.sort_index()

# Plot Line Chart
plt.figure(figsize=(14, 6))
for brand in pivot_sales.columns:
    plt.plot(pivot_sales.index, pivot_sales[brand], marker='o', label=brand)
plt.title('Total Units Sold Per Quarter (Top 5 Brands)')
plt.xlabel('Quarter')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.grid(True)
# Add Labels
for brand in pivot_sales.columns:
    for x, y in zip(pivot_sales.index, pivot_sales[brand]):
        plt.text(x, y + 1, f"{int(y):,}", ha='center', va='bottom', fontsize=8)
plt.legend(title="Brand")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Units Sold Per Month By Customer Segment, Vehicle Type, and Fast Charging Option
data = pd.read_csv("train.csv")

# Convert Date column
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Extract Year-Month
data['YearMonth'] = data['Date'].dt.to_period('M').astype(str)

# Convert to numeric
data['Units_Sold'] = pd.to_numeric(data['Units_Sold'], errors='coerce')

# Remove NaN
data = data.dropna(subset=['Units_Sold'])

# Group by categories
group_cols = ['Customer_Segment', 'Vehicle_Type', 'Fast_Charging_Option']

for col in group_cols:
    # Group by category and YearMonth
    monthly_sales = data.groupby([col, 'YearMonth'])['Units_Sold'].sum().reset_index()

    # Pivot table for multi-line chart
    pivot_sales = monthly_sales.pivot(index='YearMonth', columns=col, values='Units_Sold')

    # Sort Month order
    pivot_sales = pivot_sales.sort_index()

    # Plot Line Chart
    plt.figure(figsize=(14, 6))
    for category in pivot_sales.columns:
        plt.plot(pivot_sales.index, pivot_sales[category], marker='o', label=category)
    plt.title(f'Total Units Sold Per Month (By {col})')
    plt.xlabel('Month')
    plt.ylabel('Total Units Sold')
    plt.xticks(rotation=45)
    plt.grid(True)
    # Add Labels
    for category in pivot_sales.columns:
        for x, y in zip(pivot_sales.index, pivot_sales[category]):
            if pd.notna(y):
                plt.text(x, y + 1, f"{int(y):,}", ha='center', va='bottom', fontsize=8)
    plt.legend(title=col)
    plt.tight_layout()
    plt.show()
```
