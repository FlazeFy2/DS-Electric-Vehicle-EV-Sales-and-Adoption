```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Unit Sold Per Month in 2023
data = pd.read_csv("train.csv")

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['YearMonth'] = data['Date'].dt.to_period('M').astype(str)

# Group by YearMonth and sum Unit Sold
monthly_sales = data.groupby('YearMonth')['Units_Sold'].sum()

# Plot Line Chart
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Unit Sold Per Month in 2023')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.grid(True)
# Add Labels
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y + 1, str(int(y)), ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Revenue Per Month in 2023
data = pd.read_csv("train.csv")

data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['YearMonth'] = data['Date'].dt.to_period('M').astype(str)

# Group by YearMonth and sum Unit Sold
monthly_sales = data.groupby('YearMonth')['Revenue'].sum()

# Plot Line Chart
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Revenue Per Month in 2023')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(True)
# Add Labels
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y + 1, f"{int(y):,}", ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
```
